```python
#2018 floor
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
p1_2018=pd.read_csv("2018Floor1.csv")
p1_2019=pd.read_csv("2019Floor1.csv")
p1_2018.head()
```

```python
P1= pd.concat([p1_2018, p1_2019], ignore_index=True)
column_name_map = {
    'z1_Light(kW)': 'Zona 1 - Iluminación (kW)',
    'z1_Plug(kW)': 'Zona 1 - Enchufes (kW)',
    'z2_AC1(kW)': 'Zona 2 - Aire Acondicionado 1 (kW)',
    'z2_AC2(kW)': 'Zona 2 - Aire Acondicionado 2 (kW)',
    'z2_AC3(kW)': 'Zona 2 - Aire Acondicionado 3 (kW)',
    'z2_AC4(kW)': 'Zona 2 - Aire Acondicionado 4 (kW)',
    'z2_Light(kW)': 'Zona 2 - Iluminación (kW)',
    'z2_Plug(kW)': 'Zona 2 - Enchufes (kW)',
    'z3_Light(kW)': 'Zona 3 - Iluminación (kW)',
    'z3_Plug(kW)': 'Zona 3 - Enchufes (kW)',
    'z4_Light(kW)': 'Zona 4 - Iluminación (kW)'
}
P1 = P1.rename(columns=column_name_map)
Z1=P1[['Date','Zona 1 - Iluminación (kW)','Zona 1 - Enchufes (kW)']]
Z3=P1[['Date','Zona 3 - Iluminación (kW)','Zona 3 - Enchufes (kW)']]
Z4=P1[['Date','Zona 4 - Iluminación (kW)']]
Z2=P1[['Date','Zona 2 - Aire Acondicionado 1 (kW)','Zona 2 - Aire Acondicionado 2 (kW)','Zona 2 - Aire Acondicionado 3 (kW)','Zona 2 - Aire Acondicionado 4 (kW)','Zona 2 - Iluminación (kW)','Zona 2 - Enchufes (kW)']]

Z1.head(86000)
```

```python
def MeanHour(Z1):
    # Asegúrate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['hour'] = Z['Date'].dt.hour
    
    # Agrupar por hora y calcular el promedio
    hourly_avg = Z.groupby('hour').mean(numeric_only=True)
    hourly_des=Z.groupby('hour').std(numeric_only=True)
    numeric_cols = hourly_avg.select_dtypes(include=np.number).columns
    hourly_var=hourly_des[numeric_cols]/hourly_avg[numeric_cols]
    
    # Mostrar resultado
    return[hourly_avg,hourly_var]
def MeanDayofWeek(Z1):
     # Asegúrate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['dayofweek'] = Z['Date'].dt.dayofweek
    
    # Agrupar por hora y calcular el promedio
    dayofweek_avg = Z.groupby('dayofweek').mean(numeric_only=True)
    dayofweek_des = Z.groupby('dayofweek').std(numeric_only=True)
    numeric_cols = dayofweek_avg.select_dtypes(include=np.number).columns
    dayofweek_var=dayofweek_des[numeric_cols]/dayofweek_avg[numeric_cols]
    
    # Mostrar resultado
    return[dayofweek_avg,dayofweek_var]
def MeanMouth(Z1):
     # Asegúrate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['month'] = Z['Date'].dt.month
    
    # Agrupar por hora y calcular el promedio
    month_avg = Z.groupby('month').mean(numeric_only=True)
    
    
    # Mostrar resultado
    return(month_avg)
```

```python
#Promedios
MeanZ1=[MeanHour(Z1)[0],MeanDayofWeek(Z1)[0],MeanMouth(Z1)]
MeanZ2=[MeanHour(Z2)[0],MeanDayofWeek(Z1)[0],MeanMouth(Z2)]
MeanZ3=[MeanHour(Z3)[0],MeanDayofWeek(Z3)[0],MeanMouth(Z3)]
MeanZ4=[MeanHour(Z4)[0],MeanDayofWeek(Z4)[0],MeanMouth(Z4)]
#variabilidad
varZ1=[MeanHour(Z1)[1],MeanDayofWeek(Z1)[1]]
varZ2=[MeanHour(Z2)[1],MeanDayofWeek(Z1)[1]]
varZ3=[MeanHour(Z3)[1],MeanDayofWeek(Z3)[1]]
varZ4=[MeanHour(Z4)[1],MeanDayofWeek(Z4)[1]]

```

```python
import matplotlib.pyplot as plt
def Plotting(Z,title,xlabel,ylabel,xticks,Xlabels):
    for name in Z.columns[0::]:
        Z[name].plot(figsize=(13,6))
        plt.title(f'{title} {name}')
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True)
        plt.xticks(ticks=xticks,labels=Xlabels)
        plt.show()
        
def MainPlotting(Z,flag=True):
    
    MeanHourZ=Z[0]
    MeanDayofWeekZ=Z[1]
    
    dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
    Plotting(MeanHourZ,'Consumo promedio por hora del día de:','Hora del día','Consumo promedio (kW)',range(0,24),range(0,24))
    Plotting(MeanDayofWeekZ,'Consumo promedio por día de la semana de:','dia de la semana','Consumo promedio (kW)',range(0,7),dias)
    if flag:
        meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
         'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
        MeanMouthZ=Z[2]
        Plotting(MeanMouthZ,'Consumo promedio por mes del año de:','mes del año','Consumo promedio (kW)',range(1,13),meses)
    
    
```
