```python
#2018 floor
import pandas as pd
p1_2018=pd.read_csv("2018Floor1.csv")
p1_2019=pd.read_csv("2019Floor1.csv")
p1_2018.head()
```

```python
P1= pd.concat([p1_2018, p1_2019], ignore_index=True)
Z1=P1[['Date','z1_Light(kW)','z1_Plug(kW)']]
Z3=P1[['Date','z3_Light(kW)','z3_Plug(kW)']]
Z4=P1[['Date','z4_Light(kW)']]
Z2=P1[['Date','z2_AC1(kW)','z2_AC2(kW)','z2_AC3(kW)','z2_AC4(kW)']]

Z1.head(86000)
```

```python
def MeanHour(Z1):
    # Aseg√∫rate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['hour'] = Z['Date'].dt.hour
    
    # Agrupar por hora y calcular el promedio
    hourly_avg = Z.groupby('hour').mean()
    
    # Mostrar resultado
    return(hourly_avg)
def MeanDayofWeek(Z1):
     # Aseg√∫rate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['dayofweek'] = Z['Date'].dt.dayofweek
    
    # Agrupar por hora y calcular el promedio
    dayofweek_avg = Z.groupby('dayofweek').mean()
    
    # Mostrar resultado
    return(dayofweek_avg)
def MeanMouth(Z1):
     # Aseg√∫rate de que 'Date' es datetime
    Z=Z1.copy()
    Z['Date'] = pd.to_datetime(Z['Date'])
    
    # Extraer la hora
    Z['month'] = Z['Date'].dt.month
    
    # Agrupar por hora y calcular el promedio
    month_avg = Z.groupby('month').mean()
    
    # Mostrar resultado
    return(month_avg)
```

```python
MeanZ1=[MeanHour(Z1),MeanDayofWeek(Z1),MeanMouth(Z1)]
MeanZ2=[MeanHour(Z2),MeanDayofWeek(Z1),MeanMouth(Z2)]
MeanZ3=[MeanHour(Z3),MeanDayofWeek(Z3),MeanMouth(Z3)]
MeanZ4=[MeanHour(Z4),MeanDayofWeek(Z4),MeanMouth(Z4)]
```

```python
import matplotlib.pyplot as plt

MeanZ1[0]['z1_Plug(kW)'].plot(figsize=(10,6))
plt.title('Consumo promedio por hora del d√≠a')
plt.xlabel('Hora del d√≠a')
plt.ylabel('Consumo promedio (kW)')
plt.grid(True)
plt.xticks(range(0,24))
plt.show()
```
